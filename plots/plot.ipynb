```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
!pwd
```

```python
auc_baseline = [0.8699254349627175, 0.8024300895469847, 0.6383507620302971]
auc_esm1b_whole = [0.867439933719966, 0.7996652072194964, 0.6316254395624535]
auc_msa_whole = [0.8624689312344657, 0.7863706702456126, 0.614216933043393]
auc_esm1b_separated = [0.8525269262634632, 0.8152442557186126, 0.616165148830451]
auc_msa_separated = [0.844241922120961, 0.7831034153997345, 0.5768123634972433]

ACC_list = list()
ACC_list.append(auc_baseline[0])
ACC_list.append(auc_esm1b_whole[0])
ACC_list.append(auc_msa_whole[0])
ACC_list.append(auc_esm1b_separated[0])
ACC_list.append(auc_msa_separated[0])


AUC_list = list()
AUC_list.append(auc_baseline[1])
AUC_list.append(auc_esm1b_whole[1])
AUC_list.append(auc_msa_whole[1])
AUC_list.append(auc_esm1b_separated[1])
AUC_list.append(auc_msa_separated[1])


MCC_list = list()
MCC_list.append(auc_baseline[2])
MCC_list.append(auc_esm1b_whole[2])
MCC_list.append(auc_msa_whole[2])
MCC_list.append(auc_esm1b_separated[2])
MCC_list.append(auc_msa_separated[2])

```

```python
# plot data in grouped manner of bar type
params = {'legend.fontsize': 15,
          'legend.handlelength': 1}
plt.rcParams.update(params)

plt.rcParams["figure.figsize"] = (20,12)
x = np.arange(5)
width = 0.15
metrics_list = ['ACC', 'AUC', 'MCC']

plt.plot(x-0.15, ACC_list, "s", label="ACC".format("s"), markersize=40, color = "skyblue")


plt.plot(x, AUC_list, "<", label="AUC".format("<"), markersize=40, color = "plum")


plt.plot(x+0.15, MCC_list, "o", label="MCC".format("o"), markersize=40, color = "seagreen")

legend_list = ["Baseline", "ESM-1b (whole)", "ESM-MSA-1b (whole)", "ESM-1b (separated)", "ESM-MSA-1b (separated)"]

plt.xticks(x, legend_list, size=18)
plt.yticks(size=18)
plt.title("ACC, AUC and MCC Values of Different Encoding Methods", size=24)
plt.xlabel("Encoding Method", size=23)
plt.ylabel("Metrics", size=23)
plt.legend(metrics_list, loc="upper right", fontsize= 10)

lgnd = plt.legend(loc="upper right", numpoints=1, fontsize=15)

#change the marker size manually for both lines
lgnd.legendHandles[0]._legmarker.set_markersize(17)
lgnd.legendHandles[1]._legmarker.set_markersize(17)
lgnd.legendHandles[2]._legmarker.set_markersize(17)



for t, m, text in zip(x, ACC_list, ACC_list):
    plt.text(t-0.22, m-0.018, round(text,2), fontsize=15, color="gray")

for t, m, text in zip(x, AUC_list, AUC_list):
    plt.text(t-0.05, m-0.0165, round(text,2), fontsize=15, color="gray")

c = 0
for t, m, text in zip(x, MCC_list, MCC_list):
    c += 1
    if c == 5:
        break
    plt.text(t+0.08, m-0.018, round(text,2), fontsize=15, color="gray")
    

c = 0
for t, m, text in zip(x, MCC_list, MCC_list):
    c += 1
    if c == 5:
        plt.text(t+0.08, m+0.0135, round(text,2), fontsize=15, color="gray")    
plt.show()
```
